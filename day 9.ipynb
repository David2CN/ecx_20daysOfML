```python
import numpy as np
import pandas as pd
headers = ("bkblk,bknwy,bkon8,bkona,bkspr,bkxbq,bkxcr,bkxwp,blxwp,bxqsq,cntxt,dsopp,dwipd, hdchk,katri,mulch,qxmsq,r2ar8,reskd,reskr,rimmx,rkxwp,rxmsq,simpl,skach,skewr,skrxp,spcop,stlmt,thrsk,wkcti,wkna8,wknck,wkovl,wkpos,wtoeg,target").split(",")
chess = pd.read_csv("krkp.csv")
chess.columns = headers
chess.head()
```

```python
def data_split(X, y, test_size = 0.3):
    """
    To split the data into train and test.
    """
    size = int(test_size * X.shape[0])
    X_train = X[size : , : ]
    X_test = X[0 : size , : ]
    y_train = y[size : ]
    y_test = y[0 : size ]
    return X_train, X_test, y_train, y_test

features = chess.iloc[ : , 0 : 36]
head1 =
target = chess["target"]
#70 - 30 split as required
X_train, X_test, y_train, y_test = data_split(features.values, target.values)
```

```python
#Preprocessing: encoding the features.
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
le = LabelEncoder()
target =  le.fit_transform(target)
oe = OrdinalEncoder()
head1 = features.columns
features = oe.fit_transform(features)
features = pd.DataFrame(features, columns = head1)
target = pd.DataFrame(target, columns = ["target"])
```

```python
temp = pd.concat([features, target], axis = 1)
correlations = temp.corr()
res = correlations["target"].abs().sort_values()
#To extract the top 18 most correlated features with the target.
distribution = res.quantile([19/37])
cut_off = distribution[19/37]
map = res[res >= cut_off]
top18 = temp[map.keys()]
```
